## Experiment 1.2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

```python
sigma = [0.5, 1, 2, 4, 8]
tot_points = 100

center_a = np.array([-1,-1])
center_b = np.array([1,-1])
center_c = np.array([0,1])

cluster_centers = np.concatenate([[center_a], [center_b], [center_c]])

labels = np.array([0]*tot_points + [1]*tot_points +[2]*tot_points)
```

```python
def get_multivariate_data(sigma):
    x_a = np.random.multivariate_normal(center_a, sigma * np.array([2, 0.5, 0.5, 1]).reshape((2,2)), tot_points)
    x_b = np.random.multivariate_normal(center_b, sigma * np.array([1, -0.5, -0.5, 2]).reshape((2,2)), tot_points)
    x_c = np.random.multivariate_normal(center_c, sigma * np.array([1, 0, 0, 2]).reshape((2,2)), tot_points)
    X = np.concatenate([x_a, x_b, x_c])
    
    return X
```

```python
out = get_multivariate_data(0.5)
map_points = {0:'r', 1:'g', 2:'b'}
plt.title("Points")
plt.scatter(out[:, 0], out[:, 1], c=[map_points[i] for i in labels])
plt.show();
```

```python
def pick_labels(pred_clusters, pred_labels):
    o_pred_cc = []
    o_pred_labels = []
    label_structure = {}
    for cci, cc in enumerate(cluster_centers):
        label_i = np.argmin(abs(cc - pred_clusters).sum(axis=1))
        label_structure[label_i] = cci
        o_pred_cc.append(pred_clusters[label_i])
        
    try: 
        o_pred_labels = [label_structure[i] for i in pred_labels]
    except KeyError:
        o_pred_labels = [0]*300
    return o_pred_cc, o_pred_labels

def get_metrics(labels, pred_labels, pred_centers, X):  
    obj_function = 0
    for i, j in zip(pred_labels, X):
        obj_function += np.linalg.norm(pred_centers[i] - j)**2
    summ=0
    for i in range(0,300):
        if labels[i]==pred_labels[i]:
            summ+=1
    accuracy = summ/300
    return accuracy, obj_function
     
```

```python
def getcluster(model_name):

    obj_f_s = []
    acc_s = []

    for sig in sigma:

        best_obj_function = np.inf
        best_acc = np.inf
        for _ in range(5):
            X = get_multivariate_data(sig) 
            if model_name == 'gm':
                gm = GaussianMixture(n_components=3).fit(X)
                gm_means=gm.means_
                pred_cc, pred_labels = pick_labels(gm_means, gm.predict(X))
            elif model_name == 'kmeans':
                kmeans = KMeans(n_clusters=3, n_init=10).fit(X)
                cluster_centers=kmeans.cluster_centers_
                labelsk=kmeans.labels_
                pred_cc, pred_labels = pick_labels(cluster_centers, labelsk)
            accuracy, obj_function = get_metrics(labels, pred_labels, pred_cc, X)
                
            if obj_function < best_obj_function:
                best_accuracy = accuracy
                best_obj_function = obj_function
                
                
        
        obj_f_s.append(best_obj_function)
        acc_s.append(best_accuracy)
        
    return obj_f_s, acc_s
```

```python
k_obj_f_s, k_acc_s = getcluster('kmeans')
gm_obj_f_s, gm_acc_s = getcluster('gm')
plt.plot(sigma, k_obj_f_s, marker='o', label = 'kmeans')
plt.plot(sigma, gm_obj_f_s, marker='o', label = 'gm')
plt.xlabel('sigma')

plt.ylabel('objective function')
plt.legend()
plt.title("Objective function")
plt.savefig('images/knn_objective.jpg')
plt.show();
```

```python
plt.plot(sigma, k_acc_s, marker='o', label = 'kmeans')
plt.plot(sigma, gm_acc_s, marker='o', label = 'gm')
plt.xlabel('sigma')
plt.title("Accuracy")
plt.ylabel('cluster accuracy')
plt.legend()
plt.savefig('images/knn_accuracy.jpg')
plt.show();
```

## Experiment 2.3

```python
import numpy as np
import pandas as pd
import numpy.linalg as la
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
from numpy.testing import assert_array_almost_equal
import matplotlib.pyplot as plt


np.set_printoptions(precision=3)

D2 = pd.read_csv('data/data2D.csv', header=None, index_col=None)
D1000 = pd.read_csv('data/data1000D.csv', header=None, index_col=None)

_, S, _ = np.linalg.svd((D1000 - D1000.mean()))
plt.figure(figsize=(20,5))

plt.xticks(np.arange(0, len(S)+1, 10))
plt.xlabel('Index of singular values')
plt.ylabel('Singular values')
plt.title('Singular values')
plt.plot(S)
plt.show()
```

```python
def BuggyPCA(data,d):
    
    mean = np.zeros(data.shape[1])
    std = np.ones(data.shape[1])
    title = 'Buggy PCA'
    k = d
    u, s, vt = np.linalg.svd((data - mean) / std)
    components = vt[:k]
    
    datamean=data-mean
    data_pca = ((datamean) / std).dot(components.T)
    data_projected = (data_pca.dot(components)* std) + mean
    reconstruction_error = np.sum((data - data_projected) ** 2, axis=1).mean()
    return title,reconstruction_error,data_projected

def DemeanedPCA(data,d):
    
    mean = data.mean(axis=0)
    std = std = np.ones(data.shape[1])
    title = 'Demeaned PCA'
    k = d
    u, s, vt = np.linalg.svd((data - mean) / std)
    components = vt[:k]
    
    datamean=data-mean
    data_pca = ((datamean) / std).dot(components.T)
    data_projected = (data_pca.dot(components)* std) + mean
    reconstruction_error = np.sum((data - data_projected) ** 2, axis=1).mean()
    return title,reconstruction_error,data_projected

def NormalizedPCA(data,d):
    
    mean = data.mean(axis=0)
    std = data.std(axis=0)
    title = 'Normalized PCA'
    k = d
    u, s, vt = np.linalg.svd((data - mean) / std)
    components = vt[:k]
    
    datamean=data-mean
    data_pca = ((datamean) / std).dot(components.T)
    data_projected = (data_pca.dot(components)* std) + mean
    reconstruction_error = np.sum((data - data_projected) ** 2, axis=1).mean()
    return title,reconstruction_error,data_projected

def DRO(data,d):
    data = data.to_numpy()
    components =[0, 0 ,0]
    n=data.shape[0]
    datamean= data.mean(axis=0)
    U, S, VT = np.linalg.svd((data - datamean))
    S = np.diag(S)
    k=d
    components[0] = U[:,:k]
    components[2] = VT[:k]
    components[1] = S[:k,:k]
    Z = np.sqrt(n)*components[0]
    A = ((1/np.sqrt(n))*(components[1] @ components[2])).T
    data_projected = (Z @ A.T) + datamean
    reconstruction_error = np.sum((data - data_projected) ** 2, axis=1).mean()
    title='DRO'
    return Z,A,datamean,title,reconstruction_error,data_projected

def Epochss(data, A, eta, b):
    d = A.shape[1]
    n = data.shape[0]
    D = data.shape[1]
    Xmb = (data - b.T) 
    K = ( A @ A.T ) + eta**2 * np.identity(D) 
    Kinv = np.linalg.inv(K) 
    RzxMeans = ((data - b.T) @ Kinv) @ A
    RzxVar = np.identity(d) - A.T @ Kinv @ A 
    EAZ =((Xmb @ Kinv) @ A )@ A.T 
    EAZ2 = np.sum( EAZ**2 , axis=1) + np.sum(np.diag(A @ RzxVar @ A.T)) 
    M1 = Xmb.T @ ((Xmb @ Kinv) @ A ) 
    M2 = ((Xmb @ Kinv) @ A ).T @ ((Xmb @ Kinv) @ A ) + n * RzxVar; 
    ANew = M1 @ np.linalg.inv(M2)
    x1=1/(n*D)
    x2=np.linalg.norm(Xmb,  ord='fro')
    x3=np.sum(np.sum((Xmb* EAZ)))
    etaNew =  np.sqrt(x1 * (x2**2 - 2*x3+ np.sum(EAZ2)));
    return ANew, etaNew

def DRLV(data, d=1):
    components =[0, 0 ,0]
    n = data.shape[0]
    D = data.shape[1]
    k = d
    #parameter initialization using DRO
    Z, A, b, _,_,Y = DRO(data, k)
    data = data.to_numpy()
    eta = np.sqrt(np.mean(np.mean((Y - data)**2)))
    for i in range(10):
        A, eta = Epochss(data, A, eta, b)
    Z = (data - b.T) @ (np.linalg.inv( A @ A.T + eta**2 * np.identity(D)) @ A) ;
    data_projected = (Z @ A.T) + b
    reconstruction_error = np.sum((data - data_projected) ** 2, axis=1).mean()
    title="DRLV"
    return title,reconstruction_error,data_projected

def PCA(data, d=1, PCAtype=0,plot=True):
    if(PCAtype == 0):
        print("\nResults for Buggy PCA\n")
        title,reconstruction_error,data_projected=BuggyPCA(data,d)
    elif(PCAtype == 1):
        print("\nResults for Demeaned PCA\n")
        title,reconstruction_error,data_projected=DemeanedPCA(data,d)
    elif(PCAtype == 2):
        print("\nResults for Normalized PCA\n")
        title,reconstruction_error,data_projected=NormalizedPCA(data,d)
    elif(PCAtype == 3):
        print("\nResults for DRO\n")
        Z,A,datamean,title,reconstruction_error,data_projected=DRO(data,d)
    elif(PCAtype == 4):
        print("\nResults for DRLV\n")
        title,reconstruction_error,data_projected=DRLV(data,d)
    
    print("Reconstruction Error : ", reconstruction_error)
    if (plot):
        plt.figure(figsize = (4,4))
        plt.scatter(data[0].to_numpy(), data[1].to_numpy(), marker='*' )
        if(PCAtype==3 or PCAtype==4):
            plt.scatter(data_projected[:,0], data_projected[:,1], marker='*' )
        else:    
            plt.scatter(data_projected[0].to_numpy(), data_projected[1].to_numpy(), marker='*' )
        plt.xlabel('X1')
        plt.ylabel('X2')
        plt.title(title)
```

```python
for i in range(5):
    PCA(D2, d=1, PCAtype=i, plot=True)

```
